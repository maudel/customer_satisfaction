## 4 · Paso 6 · Limpieza de Variables

`src/cleaning.py::LimpiadorDatos` implementa una interfaz **ajustar/transformar**: los parámetros
(cortes de clipado, medianas de imputación, categorías frecuentes) se **aprenden solo del set de TRAIN**
y se aplican idénticos a las demás particiones. Esto **evita fuga de información** y garantiza que el
mismo tratamiento se replique mes a mes.

- **Clipado (winsorización)** de continuas a los percentiles 1-99.
- **Imputación**: numéricas → mediana; binarias → 0; categóricas → `OTHER`.
- **Agrupamiento** de categorías con frecuencia < 1% en la etiqueta `OTHER`.

````python
limpiador = cleaning.LimpiadorDatos().ajustar(train)
train_limpio = limpiador.transformar(train)
val_limpio   = limpiador.transformar(val)
print("Reporte de limpieza (parámetros aprendidos en TRAIN):\n")
limpiador.reporte()
````

## 5 · Paso 7 · Selección de Variables (cascada del flujo de trabajo)

`src/selection.py::SelectorVariables` ejecuta la cascada **en el orden exacto del Excel**, evaluando
cada paso con un **Random Forest** y midiendo **AUC-ROC en train y val**:

| Orden | Método | Qué descarta |
|-------|--------|--------------|
| 0 | Estado Inicial | — |
| 1 | `metodo_faltantes` | features con > `umbral_faltantes` de nulos |
| 2 | `metodo_psi` | features **inestables en el tiempo** (PSI train↔val > umbral) |
| 3 | `metodo_correlacion` | features **redundantes** entre sí (\|corr\| > umbral) |
| 4 | `metodo_univariante` | features **sin señal** vs target (Gini univariante < umbral) |
| 5 | `WOE / Information Value` | *ranking* complementario de poder predictivo |

Umbrales (en `config.py`): `umbral_faltantes=10%`, `bins_PSI=10`, `umbral_correlacion=0.95`, `umbral_univariante=0.05`.

````python
selector = selection.SelectorVariables().ajustar(train_limpio, val_limpio)
print("Cascada de selección (igual que la tabla del flujo de trabajo):\n")
selector.tabla_reporte()
````

````python
# Tabla de Information Value (WOE) — poder predictivo univariante
print("Information Value de las features numéricas/binarias:\n")
selector.tabla_iv()
````

````python
variables = selector.variables_seleccionadas_
print(f"VARIABLES SELECCIONADAS PARA EL MODELO ({len(variables)}):\n")
for v in variables:
    print("  •", v)

# Guardar reporte de selección
selector.tabla_reporte().to_csv(C.DIR_REPORTES / "feature_selection_report.csv", index=False)
print("\n[Reporte guardado en reports/feature_selection_report.csv]")
````

````python
# Visualización de la cascada: features eliminadas vs AUC conservada
reporte = selector.tabla_reporte()
fig, ax1 = plt.subplots(figsize=(11, 4.8))
ax1.bar(reporte["metodo"], reporte["features"], color="#85C1E9", edgecolor="white", alpha=0.9)
ax1.set_ylabel("Nº de features", color="#1B4F72"); ax1.set_xlabel("Paso de la cascada")
ax1.tick_params(axis="x", rotation=20)
for i, v in enumerate(reporte["features"]):
    ax1.text(i, v + 0.3, str(v), ha="center", fontweight="bold")

ax2 = ax1.twinx()
ax2.plot(reporte["metodo"], reporte["auc_roc_train"], "o-", color="#1B4F72", label="AUC train")
ax2.plot(reporte["metodo"], reporte["auc_roc_val"], "s--", color="#E74C3C", label="AUC val")
ax2.set_ylabel("AUC-ROC"); ax2.set_ylim(0.5, 0.8); ax2.legend(loc="lower left")
ax1.set_title("Selección de variables: reducción de features vs AUC conservada",
              fontweight="bold")
plt.tight_layout(); plt.savefig(C.DIR_FIGURAS / "s2_fig02_feature_selection.png", dpi=150, bbox_inches="tight")
plt.show()
````

**Barrido del umbral univariante** (igual que el Excel: `umbral_univariante = [0.1, 0.2, 0.3]`).

````python
filas = []
for u in C.UMBRAL_UNIVARIANTE_BARRIDO:
    s = selection.SelectorVariables(umbral_univariante=u).ajustar(train_limpio, val_limpio)
    auc_train, auc_val = selection.evaluar_variables(train_limpio, val_limpio, s.variables_seleccionadas_)
    filas.append({"umbral_univariante": u, "n_features": len(s.variables_seleccionadas_),
                  "auc_train": round(auc_train, 4), "auc_val": round(auc_val, 4)})
pd.DataFrame(filas)
````

## 6 · Pasos 5-6 · Pipeline Reproducible (scikit-learn)

`src/pipeline.py` ensambla un `Pipeline` de scikit-learn que encapsula **todo** el preprocesamiento
+ el modelo en un único objeto serializable:

```
ColumnTransformer
 ├─ numéricas   → SimpleImputer(mediana) → StandardScaler
 ├─ binarias    → SimpleImputer(moda)
 └─ categóricas → SimpleImputer("OTHER") → OneHotEncoder(handle_unknown="ignore")
        ↓
RandomForestClassifier   (baseline del Sprint 2; se optimiza en Sprint 3)
```

El `handle_unknown="ignore"` del encoder garantiza que **categorías nuevas** que aparezcan en meses
futuros no rompan la predicción — clave para la operación mensual.

````python
pipe = pipeline.construir_pipeline(variables)
pipe.fit(train_limpio, train_limpio[C.OBJETIVO])
print("Pipeline entrenado.")
pipe
````

## 7 · Métricas Técnicas y de Negocio Finales (versión pipeline)

Se evalúa el pipeline completo sobre **val**, **backtest** y **live**.
- **Técnicas:** Accuracy, Precision, Recall, F1, ROC-AUC, **Gini** (= 2·AUC−1).
- **Negocio:** Customer Satisfaction Rate, Revenue at Risk, y **% de insatisfechos detectados**
  (utilidad del modelo para intervención temprana).

````python
resultados_tec, resultados_neg = {}, {}
for nombre_split in ["val", "backtest", "live"]:
    particion = datos[datos.dataset_split == nombre_split].copy()
    if particion.empty:
        continue
    particion_limpia = limpiador.transformar(particion)
    proba = pipe.predict_proba(particion_limpia[variables])[:, 1]
    resultados_tec[nombre_split] = metrics.metricas_tecnicas(particion_limpia[C.OBJETIVO], proba)
    resultados_neg[nombre_split] = metrics.metricas_negocio(particion_limpia, proba)

print("MÉTRICAS TÉCNICAS:")
display(pd.DataFrame(resultados_tec).T)
print("\nMÉTRICAS DE NEGOCIO:")
pd.DataFrame(resultados_neg).T
````

````python
# Matriz de confusión sobre backtest
backtest = datos[datos.dataset_split == "backtest"].copy()
backtest_limpio = limpiador.transformar(backtest)
proba_bt = pipe.predict_proba(backtest_limpio[variables])[:, 1]
print("Matriz de confusión (BACKTEST):")
metrics.matriz_confusion(backtest_limpio[C.OBJETIVO], proba_bt)
````

## 8 · Simulación de Incorporación Mensual de Datos

Se simula la llegada de datos **mes a mes** (backtest + live): para cada mes se puntúa con el pipeline
ya entrenado y se registran las métricas. Esto valida que el modelo **se mantiene estable en el tiempo**
y prepara el *retraining* mensual automatizado (paso "Aplicar el Pickle" del flujo de trabajo).

````python
filas_mensuales = []
for mes in C.MESES_BACKTEST + [C.MES_PRODUCCION]:
    particion = datos[datos[C.COL_MES] == mes].copy()
    if particion.empty:
        continue
    particion_limpia = limpiador.transformar(particion)
    proba = pipe.predict_proba(particion_limpia[variables])[:, 1]
    mt = metrics.metricas_tecnicas(particion_limpia[C.OBJETIVO], proba)
    filas_mensuales.append({"mes": mes, "n": mt["n"], "roc_auc": mt["roc_auc"],
                            "f1": mt["f1"], "gini": mt["gini"]})
df_mensual = pd.DataFrame(filas_mensuales)
df_mensual.to_csv(C.DIR_REPORTES / "monthly_backtest.csv", index=False)
df_mensual
````

````python
fig, ax = plt.subplots(figsize=(10, 4.5))
ax.plot(df_mensual["mes"], df_mensual["roc_auc"], "o-", color="#1B4F72", lw=2, label="ROC-AUC")
ax.plot(df_mensual["mes"], df_mensual["gini"], "s--", color="#E74C3C", lw=2, label="Gini")
ax.plot(df_mensual["mes"], df_mensual["f1"], "^-", color="#2ECC71", lw=2, label="F1")
ax.set_title("Estabilidad del modelo mes a mes (backtest out-of-time)", fontweight="bold")
ax.set_xlabel("Mes"); ax.set_ylabel("Métrica"); ax.set_ylim(0, 1); ax.legend()
plt.tight_layout(); plt.savefig(C.DIR_FIGURAS / "s2_fig03_monthly_stability.png", dpi=150, bbox_inches="tight")
plt.show()
````

## 9 · Aplicar el Pickle a un Mes Nuevo (predicción sin label)

Se serializa el pipeline + features + metadatos en un **único `.pkl` versionado** y se aplica al mes
de predicción (`2018-09`), que simula datos nuevos **sin label**. Es exactamente el flujo de producción mensual.

````python
meta = utils.metadatos_ejecucion({"n_features": len(variables), "features": variables,
                                  "stage": "sprint2_baseline"})
ruta_artefacto = pipeline.guardar_pipeline(pipe, variables, meta)

# Recargar y aplicar a datos nuevos
artefacto = pipeline.cargar_pipeline(ruta_artefacto)
mes_pred = datos[datos.dataset_split == "predict"].copy()
mes_pred_limpio = limpiador.transformar(mes_pred)
puntuados = pipeline.predecir_datos_nuevos(artefacto, mes_pred_limpio)

print(f"Mes de predicción: {C.MES_PREDICCION} · {len(puntuados):,} pedidos puntuados")
print(f"Predichos como satisfechos:  {puntuados['pred_satisfecho'].mean()*100:.1f}%")
print(f"Pedidos en riesgo (alerta):  {(puntuados['pred_satisfecho']==0).sum():,}")
puntuados[["order_id", "delivery_delay_days", "has_comment",
           "proba_satisfecho", "pred_satisfecho"]].head(10)
````

````python
# Guardar métricas finales del Sprint 2 (versión pipeline)
metricas_finales = {**meta, "metrics": {"technical": resultados_tec,
                                        "business": resultados_neg,
                                        "monthly_backtest": filas_mensuales}}
utils.escribir_json(metricas_finales, C.DIR_REPORTES / "metrics_sprint2.json")
print("Métricas finales guardadas en reports/metrics_sprint2.json")
print("Versión del pipeline:", C.VERSION_PIPELINE)
````
